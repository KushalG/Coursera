```python
import graphlab
products = graphlab.SFrame('amazon_baby.gl')
```

```python
products.head()
```

```python
selected_words = ['awesome','great','fantastic','amazing','love','horrible','bad','terrible','awful','wow','hate']
```

```python
selected_words
```

```python
products['word count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python
for word in selected_words:
    products[word] = products['word count'].apply(lambda counts: counts.get(word,0))
```
